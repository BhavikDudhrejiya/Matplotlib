```python
#Importing library
import matplotlib.pyplot as plt
import numpy as np
import random
```

```python
#Creating data
ml_students_age = np.random.randint(18,45,(100))
py_students_age = np.random.randint(15,40,(100))

print(ml_students_age)
print(py_students_age)
```

```python
#Creating Histogram
plt.hist(ml_students_age)
plt.title('Machine Learning Students Age')
plt.xlabel('Students Age Category')
plt.ylabel('No. Students age')
plt.show()
```

```python
#Creating Histogram with bins
bins = [15,20,25,30,35,40,45]
plt.hist(ml_students_age, bins)
plt.title('Machine Learning Students Age')
plt.xlabel('Students Age Category')
plt.ylabel('No. Students age')
plt.show()
```

```python
#Due to all bar are joined together We are unable to make proper analysis hence We will create space between the bars
```

```python
#Creating Histogram with bins and bar width
bins = [15,20,25,30,35,40,45]
plt.hist(ml_students_age, bins, rwidth=0.8)
plt.title('Machine Learning Students Age')
plt.xlabel('Students Age Category')
plt.ylabel('No. Students age')
plt.show()
```

```python
#Creating Histogram with hist type
bins = [15,20,25,30,35,40,45]
plt.hist(ml_students_age, bins, rwidth=0.8, histtype='step') #{'bar', 'barstacked', 'step', 'stepfilled'}
plt.title('Machine Learning Students Age')
plt.xlabel('Students Age Category')
plt.ylabel('No. Students age')
plt.show()
```

```python
#Creating Histogram with changing orientation
bins = [15,20,25,30,35,40,45]
plt.hist(ml_students_age, bins, rwidth=0.8, histtype='bar', orientation='horizontal') #{'bar', 'barstacked', 'step', 'stepfilled'}
plt.title('Machine Learning Students Age')
plt.xlabel('Students Age Category')
plt.ylabel('No. Students age')
plt.show()
```

```python
#Creating Histogram with changing bar color
bins = [15,20,25,30,35,40,45]
plt.hist(ml_students_age, bins, rwidth=0.8, histtype='bar', 
         orientation='horizontal', color = 'm') #{'bar', 'barstacked', 'step', 'stepfilled'}
plt.title('Machine Learning Students Age')
plt.xlabel('Students Age Category')
plt.ylabel('No. Students age')
plt.show()
```

```python
#Creating Histogram with changing bar color
bins = [15,20,25,30,35,40,45]
plt.hist(ml_students_age, bins, rwidth=0.8, histtype='bar', 
         orientation='horizontal', color = 'm', label='ML Student') #{'bar', 'barstacked', 'step', 'stepfilled'}
plt.title('Machine Learning Students Age')
plt.xlabel('Students Age Category')
plt.ylabel('No. Students age')
plt.legend()
plt.show()
```

```python
#Creating Histogram
bins = [15,20,25,30,35,40,45]
plt.hist(ml_students_age, bins, rwidth=0.8, histtype='bar', 
         orientation='vertical', color = 'm', label='ML Student') #{'bar', 'barstacked', 'step', 'stepfilled'}
plt.hist(py_students_age, bins, rwidth=0.8, histtype='bar', 
         orientation='vertical', color = 'y', label='PY Student')
plt.title('Machine Learning Students Age')
plt.xlabel('Students Age Category')
plt.ylabel('No. Students age')
plt.legend()
plt.show()

#We are unable to visualize plot, We need side by side plots
```

```python
#Creating Histogram with changing bar color
bins = [15,20,25,30,35,40,45]
plt.hist([ml_students_age,py_students_age], 
         bins, rwidth=0.8, 
         histtype='bar', 
         orientation='vertical', 
         color = ['m','y'], 
         label=['ML Student','PY Student']) #{'bar', 'barstacked', 'step', 'stepfilled'}
plt.title('Machine Learning Students Age')
plt.xlabel('Students Age Category')
plt.ylabel('No. Students age')
plt.legend(loc = 2)
plt.show()
```

```python
#Creating Histogram
plt.figure(figsize=(16,9))
bins = [15,20,25,30,35,40,45]
plt.hist([ml_students_age,py_students_age], 
         bins, rwidth=0.8, 
         histtype='bar', 
         orientation='vertical', 
         color = ['m','y'], 
         label=['ML Student','PY Student']) #{'bar', 'barstacked', 'step', 'stepfilled'}
plt.title('Machine Learning Students Age')
plt.xlabel('Students Age Category')
plt.ylabel('No. Students age')
plt.legend()
plt.show()
```

```python
from matplotlib import style

style.use('ggplot')
#Creating Histogram
plt.figure(figsize=(16,9))
bins = [15,20,25,30,35,40,45]
plt.hist([ml_students_age,py_students_age], 
         bins, rwidth=0.8, 
         histtype='bar', 
         orientation='vertical', 
         color = ['m','y'], 
         label=['ML Student','PY Student']) #{'bar', 'barstacked', 'step', 'stepfilled'}
plt.title('Machine Learning Students Age')
plt.xlabel('Students Age Category')
plt.ylabel('No. Students age')
plt.legend()
plt.show()
```

```python
from matplotlib import style

style.use('ggplot')

#Creating Histogram
plt.figure(figsize=(16,9))
bins = [15,20,25,30,35,40,45]
plt.hist([ml_students_age,py_students_age], 
         bins, rwidth=0.8, 
         histtype='bar', 
         orientation='vertical', 
         color = ['m','y'], 
         label=['ML Student','PY Student'],
         density = True) #{'bar', 'barstacked', 'step', 'stepfilled'}
plt.title('Machine Learning Students Age')
plt.xlabel('Students Age Category')
plt.ylabel('No. Students age')
plt.legend()
plt.show()
```
