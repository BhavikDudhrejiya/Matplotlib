```python
#Creating subplot with various plots
plt.figure(figsize=(35,20))

#############################################################################################################################

plt.subplot(3,2,1)
style.use('ggplot')
plt.plot(days, pune_temp,"mo--",linewidth = 3, markersize = 10, label = 'pune temp')
plt.plot(days, mumbai_temp,"yo:",linewidth = 3, markersize = 10, label = 'mumbai temp')
#plt.axis([0,15,0,50])#xmin, xmax, ymin, ymax
plt.title('Pune Temperature', fontsize = 15)
plt.xlabel('days', fontsize = 15)
plt.ylabel('temperature', fontsize = 15)
plt.legend(loc = 4) #loc range between 0 to 10
plt.grid(color='w', linestyle='-', linewidth=2)


#############################################################################################################################

plt.subplot(3,2,2)
style.use('ggplot')
#Creating Histogram
bins = [15,20,25,30,35,40,45]
plt.hist([ml_students_age,py_students_age], 
         bins, rwidth=0.8, 
         histtype='bar', 
         orientation='vertical', 
         color = ['m','y'], 
         label=['ML Student','PY Student']) #{'bar', 'barstacked', 'step', 'stepfilled'}
plt.title('Machine Learning Students Age')
plt.xlabel('Students Age Category')
plt.ylabel('No. Students age')
plt.legend()


#############################################################################################################################

plt.subplot(3,2,3)
classes_index = np.arange(len(Classes))
width = 0.2
plt.bar(classes_index, Class1_students, width, color = 'b',label = 'Class 1 Students');
plt.bar(classes_index + width, Class2_students, width, color = 'y',label = 'Class 2 Students');
plt.bar(classes_index + width + width, Class3_students, width, color = 'g',label = 'Class 3  Students');
plt.xticks(classes_index + width, Classes, rotation = 90)
plt.title('Bar Chart of IAIP Class', fontsize = 15);
plt.xlabel('Classes',  fontsize = 15);
plt.ylabel('No. of Students',  fontsize = 15);

#############################################################################################################################

plt.subplot(3,2,4)
Classes = ['Python', 'R', 'ML', 'AI', "DS"]
Classe1_students = [45, 15, 35, 25, 30]
explode = [0.1,0,0.2,0,0]
colors = ['c', 'b', 'r', 'y', 'g']
plt.pie(Classe1_students, labels=Classes,explode = explode, colors = colors, autopct='%0.1f%%', shadow=True);

#############################################################################################################################

plt.subplot(3,2,5)
plt.scatter(x,y, color = 'r', marker='*', s = 100, alpha = 0.5, linewidths=10, edgecolors='g')
plt.scatter(x,data['Installs'], color = 'y', marker='o', s = 100, alpha = 0.5, linewidths=10, edgecolors='c')
plt.title('Google Play Store Apps Scatter Plot')
plt.xlabel('Rating')
plt.ylabel('Reviews & Installs')

#############################################################################################################################

plt.subplot(3,2,6, projection = 'polar', facecolor =  'k', frameon = True)

plt.show()
```
