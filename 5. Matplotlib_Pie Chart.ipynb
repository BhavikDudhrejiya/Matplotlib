```python
Classes = ['Python', 'R', 'ML', 'AI', "DS"]
Classe1_students = [45, 15, 35, 25, 30]
explode = [0.1,0,0.2,0,0]
textprops = {'fontsize': 15}
colors = ['c', 'b', 'r', 'y', 'g']
plt.pie(Classe1_students, 
        labels=Classes,
        explode = explode, 
        colors = colors, 
        autopct='%0.1f%%', 
        shadow=True, 
        radius=1.4,
        startangle=270, 
        textprops=textprops);
```

```python
wedgeprops = {'linewidth':4, 'width':1, 'edgecolor':'k'}
textprops = {'fontsize': 15}
plt.pie(Classe1_students, 
         labels=Classes,
         explode = explode, 
         colors = colors, 
         autopct='%0.1f%%',
         pctdistance = 0.6, #Percentage distance
         shadow=True,
         labeldistance = 1.4,
         radius=1.4, #Size of the Pie Chart 
         startangle=270, #Change the angle of pie chart
         counterclock = True, #Change the direction of the Pie Chart 
         wedgeprops= wedgeprops,
         textprops=textprops,
         center = (2,3),
         frame = True,
         rotatelabels = True);
```

```python
plt.pie(Classe1_students, 
        labels=Classes,
        explode = explode, 
        colors = colors, 
        autopct='%0.1f%%', 
        shadow=True, 
        radius=1.4,
        startangle=270, 
        textprops=textprops);
plt.legend(loc = 2)
```
