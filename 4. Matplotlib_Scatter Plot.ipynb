```python
x = data['Rating']
y = data['Reviews']
```

```python
plt.scatter(x,y)
```

```python
plt.scatter(x,y)
plt.title('Google Play Store Apps Scatter Plot')
plt.xlabel('Rating')
plt.ylabel('Reviews')
```

```python
plt.scatter(x,y, color = 'r')
plt.title('Google Play Store Apps Scatter Plot')
plt.xlabel('Rating')
plt.ylabel('Reviews')
```

```python
plt.figure(figsize=(16,9))
plt.scatter(x,y, color = 'r')
plt.title('Google Play Store Apps Scatter Plot')
plt.xlabel('Rating')
plt.ylabel('Reviews')
```

```python
plt.figure(figsize=(16,9))
plt.scatter(x,y, color = 'r', marker='*')
plt.title('Google Play Store Apps Scatter Plot')
plt.xlabel('Rating')
plt.ylabel('Reviews')
```

```python
plt.figure(figsize=(16,9))
plt.scatter(x,y, color = 'r', marker='*', s = 500)
plt.title('Google Play Store Apps Scatter Plot')
plt.xlabel('Rating')
plt.ylabel('Reviews')
```

```python
plt.figure(figsize=(16,9))
plt.scatter(x,y, color = 'r', marker='*', s = 100, alpha = 0.5)
plt.title('Google Play Store Apps Scatter Plot')
plt.xlabel('Rating')
plt.ylabel('Reviews')
```

```python
plt.figure(figsize=(16,9))
plt.scatter(x,y, color = 'r', marker='o', s = 100, alpha = 0.5, linewidths=50)
plt.title('Google Play Store Apps Scatter Plot')
plt.xlabel('Rating')
plt.ylabel('Reviews')
```

```python
plt.figure(figsize=(16,9))
plt.scatter(x,y, color = 'r', marker='*', s = 100, alpha = 0.5, linewidths=10, edgecolors='g')
plt.title('Google Play Store Apps Scatter Plot')
plt.xlabel('Rating')
plt.ylabel('Reviews')
```

```python
plt.figure(figsize=(16,9))
plt.scatter(x,y, color = 'r', s = 100, alpha = 0.5, linewidths=10, edgecolors='g', verts='<')
plt.title('Google Play Store Apps Scatter Plot')
plt.xlabel('Rating')
plt.ylabel('Reviews')
```
