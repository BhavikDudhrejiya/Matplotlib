```python
#Importin library
import matplotlib.pyplot as plt
from matplotlib import style
import numpy as np
```

```python
#Dataset
Classes = ['Python', 'R', 'AI','ML', 'DS']
Class1_students = [30, 10, 20, 25, 10]#Out of 100 each class
Class2_students = [40, 5, 20, 20, 10]
Class3_students = [35, 5, 30, 15, 15]
```

```python
#Creating Bar Plot
plt.bar(Classes, Class1_students);
```

```python
#Creating Horizontal Bar Plot
plt.barh(Classes, Class1_students);
```

```python
#Creating Bar Plot with define bar width
plt.bar(Classes, Class1_students, width = 0.2);
```

```python
#Creating Bar Plot with changing the alignment of bar
plt.bar(Classes, Class1_students, width = 0.2, align= 'edge');
```

```python
#Creating Bar Plot with changing bar color 
plt.bar(Classes, Class1_students, width = 0.2, align= 'edge', color = 'y');
```

```python
#Creating Bar Plot with changing edge bar color 
plt.bar(Classes, Class1_students, width = 0.2, align= 'edge', color = 'y',
       edgecolor = 'k');
```

```python
#Creating Bar Plot with changing edge width
plt.bar(Classes, Class1_students, width = 0.2, align= 'edge', color = 'y',
       edgecolor = 'k', linewidth = 5);
```

```python
#Creating Bar Plot with changing opacity or transperancy
plt.bar(Classes, Class1_students, width = 0.2, align= 'edge', color = 'y',
       edgecolor = 'k', linewidth = 1, alpha = 0.7);
```

```python
#Creating Bar Plot with changing edge line style
plt.bar(Classes, Class1_students, width = 0.2, align= 'edge', color = 'y',
       edgecolor = 'k', linewidth = 1, alpha = 0.7, linestyle = '--');
```

```python
#Creating Bar Plot with adding labels
plt.bar(Classes, Class1_students, width = 0.2, align= 'edge', color = 'y',
       edgecolor = 'k', linewidth = 1, alpha = 0.7, linestyle = '--', label = 'Class 1 Students');
```

```python
#Hidding Bar Plot
plt.bar(Classes, Class1_students, width = 0.2, align= 'edge', color = 'y',
       edgecolor = 'k', linewidth = 1, alpha = 0.7, linestyle = '--', label = 'Class 1 Students', visible = False);
```

```python
#Bar Plot with changing plot background
style.use('ggplot')
plt.bar(Classes, Class1_students, width = 0.6, align= 'edge', color = 'y',
       edgecolor = 'k', linewidth = 1, alpha = 0.7, linestyle = '--', label = 'Class 1 Students');
```

```python
#Bar Plot with changing plot size
style.use('ggplot')
plt.figure(figsize=(16,9))
plt.bar(Classes, Class1_students, width = 0.6, align= 'edge', color = 'y',
       edgecolor = 'k', linewidth = 1, alpha = 0.7, linestyle = '--', label = 'Class 1 Students');
```

```python
#Bar Plot with changing plot size
plt.figure(figsize=(16,9))
plt.barh(Classes, Class1_students, align= 'edge', color = 'y',
       edgecolor = 'k', linewidth = 1, alpha = 0.7, linestyle = '--', label = 'Class 1 Students');
```

```python
#Bar Plot with adding bar titles & axis labels
style.use('ggplot')
plt.figure(figsize=(16,9))
plt.bar(Classes, Class1_students, width = 0.6, align= 'edge', color = 'y',
       edgecolor = 'k', linewidth = 1, alpha = 0.7, linestyle = '--', label = 'Class 1 Students');
plt.title('Bar Chart of IAIP Class');
plt.xlabel('Classes');
plt.ylabel('No. of Students');
plt.show;
```

```python
#Bar Plot with changing font title
style.use('ggplot')
plt.figure(figsize=(16,9))
plt.bar(Classes, Class1_students, width = 0.6, align= 'edge', color = 'y',
       edgecolor = 'k', linewidth = 1, alpha = 0.7, linestyle = '--', label = 'Class 1 Students');
plt.title('Bar Chart of IAIP Class', fontsize = 15);
plt.xlabel('Classes',  fontsize = 15);
plt.ylabel('No. of Students',  fontsize = 15);
plt.show;
```

```python
#Bar Plot on two data
style.use('ggplot')
plt.figure(figsize=(16,9))

plt.bar(Classes, Class1_students, width = 0.2, align= 'edge', color = 'y',label = 'Class 1 Students');

plt.bar(Classes, Class2_students, width = 0.2, align= 'edge', color = 'g',label = 'Class 2 Students');

plt.title('Bar Chart of IAIP Class', fontsize = 15);
plt.xlabel('Classes',  fontsize = 15);
plt.ylabel('No. of Students',  fontsize = 15);
plt.show;

#Bar plots overlapping
```

```python
#Bar Plot on two data
plt.figure(figsize=(16,9))

classes_index = np.arange(len(Classes))

width = 0.2

plt.bar(classes_index, Class1_students, width, color = 'y',label = 'Class 1 Students');

plt.bar(classes_index + width, Class2_students, width, color = 'g',label = 'Class 2 Students');

plt.title('Bar Chart of IAIP Class', fontsize = 15);
plt.xlabel('Classes',  fontsize = 15);
plt.ylabel('No. of Students',  fontsize = 15);
plt.show();
```

```python
#Bar Plot on two data
plt.figure(figsize=(16,9))

classes_index = np.arange(len(Classes)) #[0, 1, 2, 3, 4]

width = 0.2

plt.bar(classes_index, 
        Class1_students, 
        width, 
        color = 'y',
        label = 'Class 1 Students');

plt.bar(classes_index + width, 
        Class2_students, 
        width, 
        color = 'g',
        label = 'Class 2 Students');

plt.xticks(classes_index + width, Classes)
plt.title('Bar Chart of IAIP Class', fontsize = 15);
plt.xlabel('Classes',  fontsize = 15);
plt.ylabel('No. of Students',  fontsize = 15);
plt.show();
```
